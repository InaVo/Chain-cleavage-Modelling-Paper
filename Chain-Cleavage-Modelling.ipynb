<h2>ABSTRACT</h2>

<p class=MsoNormal>Mechanochemical conversion of polyethylene (PE) and
polypropylene was shown to produce monomers and is thus interesting for
chemical polymer recycling. As these polymers make up more than 50% of the
worldwide polymer production, studying their conversion during ball milling is
especially relevant. However, fundamental knowledge on the effect of
crystallinity, degree of polymerization, entanglement and temperature on the
conversion is lacking due to the difficulty in producing polyolefins with
controlled chain length and dispersity. Here we synthesize PE by a controlled
chain growth polymerization and study its degradation during ball milling at
cryogenic conditions, at room temperature (RT), with and without air, and using
either steel or zirconia grinding spheres. Resulting molecular weight
distributions are fitted using a statistical chain cleavage model suggesting a statistical
Gaussian distribution of chain cleavage probability around the middle of the
chain. This is likely because if the chains fixed in an entanglement or crystal at two point they cannot slip out and force can act on them leading to cleavage. That the chain is fixed at both sides of a possible cleavage location is most likely in its center. Chain cleavage is also preferred when entangled domains exist that link crystalline regions. A micelle grown single crystal ultra-high
molecular weight PE without entanglements was milled and its molar mass
decreased much less compared to the same sample that was annealed to create
entanglements. However, in contrast to previous studies and common
expectations, the initial molar mass of the polymer, the degree of
crystallinity and the brittleness of the sample did not have a measurable
influence on chain cleavage. While the decrease in molar mass was faster at
cryogenic conditions compared to RT, nuclear magnetic resonance (NMR) results
suggest that this is due to a suppression of radical recombination rather than
the higher brittleness of the material below its glass transition temperature
(~-120&nbsp;&#730;C). Similarly, the number of permanent scissions increased by
up to 2.6 times under air compared to nitrogen atmosphere, especially in
combination with steel milling spheres. NMR spectra of the milled samples
suggest that the reaction of mechanochemically formed chains with air suppresses
recombination. </p>

## INTRODUCTION

<p class=MsoNormal>While initially being studied out of interest to prevent
degradation,<sup>1,2</sup>
mechanochemical degradation of commodity polymers is gaining attention as a
potential method for recycling.<sup>3,4</sup>
It was shown that even ball milling of polymers with a sole C-C backbone, like polymethylmethacrylate
(PMMA), polystyrene (PS),<sup>5,6</sup>
polypropylene (PP),<sup>7</sup>
and polyethylene (PE)<sup>7</sup>
can produce monomers. The collision of grinding spheres upon agitation in the
ball mill impacts polymer trapped at the collision point leading to homolytic
cleavage of the polymer chains and the formation of two radicals. Radical
formation was observed by electron spin resonance spectroscopy even during
milling in liquid N<sub>2</sub>.<sup>8,9</sup>
These radicals can undergo termination via recombination or disproportionation,
of which only the latter leads to permanent scission and a decrease in molar
mass.<sup>10</sup>
Kinetic parameters derived from thermodynamic information and rate constants of
model compounds suggests that recombination is 10 times faster than
disproportionation.<sup>11</sup>
In view of formation of monomer from the primary radicals, the desired reaction
is &#946; scission. However, for example for PP, only every 30<sup>th</sup>
radical formed was shown to release a monomer.<sup>12</sup>
</p>

<p class=MsoNormal>The initial chain cleavage rate is influenced by the polymer
morphology and the milling conditions. For fully amorphous polymers like PS and
PMMA, the glass transition temperature (<i>T<sub>g</sub></i>) was shown to
influence the cleavage rate.<sup>13</sup>
The <i>T<sub>g</sub></i> of these polymers lies above room temperature (RT). The
higher the <i>T<sub>g</sub></i>, the glassier the polymer. Cleavage seems to be
facilitated in the glassy state as impact energy cannot easily be dissipated. In
the rubbery state, polymers can disentangle upon impact instead of undergoing
chain cleavage. However, the <i>T<sub>g</sub></i> of PP and PE is far below RT.
Therefore, one would expect a higher cleavage rate at cryogenic conditions. In
addition, PE and PP are partially crystalline. The question arises as to
whether the very rigid crystalline or the amorphous regions would be more
susceptible to impact induced chain cleavage. The structure of these semicrystalline
polymers can be much more complicated (<b>Scheme 1</b>) than that of fully
amorphous materials. Usually, polymer crystals are not separated from each
other and are not perfect. Several individual chains are part of one crystal while
also protruding into amorphous regions. Polymer chains can span several
crystals as tie molecules and crystals can be connected by entanglements. However,
it was shown that when synthesized in very diluted media, single PE chains are
separated and therefore form single chain disentangled nanocrystals.<sup>14</sup>
For example, in an aqueous emulsion polymerization, single crystals of uniform
shape were synthesized with a long-lived Ni(II) catalyst.<sup>14</sup>
In this case, the crystal size depends on the molar mass, which can be tuned by
selecting a suitable polymerization time.</p>

src="Figures/Scheme 1.png"></p>

<p class=MsoNormal>The effect of the degree of crystallinity on the cleavage
rate was studied by varying the tacticity of polylactic acid (PLA). It was
found that the polymer is amorphized within minutes of milling and therefore
the initial degree of crystallinity did not impact the chain cleavage rate.<sup>15</sup>
However, for other polymers, it was found that the mechanochemical degradation
rate also depends on the initial molar mass of the polymer<sup>16</sup>
and that no further bond scission is observed for polymers below a
polymerization degree of around 100.<sup>6</sup>
It was proposed that this is because van der Waals interactions between protons
are needed to prevent the polymer chains from slipping past each other. For PE,
it was calculated that at a degree of polymerization of 83, these interactions
add up to a value close to the backbone C-C bond strength.<sup>1</sup>
Experimentally, a value between 71 and 100 was observed for mechanochemical
degradation of PE in an extruder.</p>

<p class=MsoNormal>Simulations can be used to study the cleavage location along
the length of the polymer chains. The results are often compared to degradation
kinetics or molar mass distributions. The chain cleavage probability can be
represented with different models: a) random distribution of chain cleavage
locations along the chain of the polymer, b) detachment of monomers from the chain
end c) or a chain cleavage probability of Gaussian form centered around the
middle of the chain.<sup>15</sup> It was shown that c) is the best model for PS
and PMMA cleaved during ultrasonication and ball milling by comparing
simulation results with molar mass distributions obtained from size exclusion
chromatography (SEC) after degradation of the polymers.<sup>13,17</sup>
Other authors have also simulated thermal cleavage<sup>18</sup>
and cleavage under shear in an extruder.<sup>19</sup>
However, to the best of our knowledge, none of the studies have directly fitted
experimentally obtained molar mass distributions by adjusting the parameters of
such models. </p>

<p class=MsoNormal>Whether chain cleavage during ball milling also occurs in
the middle of the chain for PP and PE is not yet known. As PE and PP are
partially crystalline in contrast to the fully amorphous PS and PMMA, studied
previously, the morphology could greatly affect the cleavage location. For
example, a tie molecule could be broken on the edge of a crystalline region
rather than in the middle of the chain. Studying the chain cleavage of these
polymers is difficult as commercial polymers usually have a rather wide molar
mass distribution (<i>M<sub>w</sub>/M<sub>n</sub></i> &gt; 3), which smears out shifts in molecular
weight.<sup>7</sup>
Modelling of the chain cleavage position is only reliable with polymers that
have a low <i>M<sub>w</sub>/M<sub>n</sub></i>. PMMA and PS
with narrow molecular weight distributions can easily be obtained by various
controlled ionic and radical polymerization techniques. In contrast, PE and PP
are mostly produced by Ziegler-Natta and Phillips, as well as, more recently
metallocene catalysts with which the molar mass distribution <i>M<sub>w</sub>/M<sub>n</sub></i>~1 cannot be achieved.<sup>20</sup>
However, there are some non-metallocene Ti based homogeneous catalysts allowing
for unprecedented control and high molar masses in ethylene solution
polymerizations.<sup>22</sup>
Weakly coordinating aryl <i>o</i>-F substituents of Ti prevent contact with the
&#946;-CH<sub>2</sub>
of the growing polymer chain, thereby suppressing &#946;-alkyl
elimination that usually leads to broad molar mass distributions.<sup>22,23</sup></p>

<p class=MsoNormal>Here we adapt a chain cleavage model previously used to
simulate the mechanochemical degradation of PS<sup>17</sup>
and star polymer<sup>24</sup>
solutions during ultrasonication to simultaneously fit molar mass distributions
obtained after various time points of ball milling. We perform the fitting for
milling under air and N<sub>2</sub>, at RT and under cryogenic conditions, as
well as, with steel and zirconia grinding spheres. To enable meaningful fits we
synthesize PE of various molar mass, all with a low <i>M<sub>w</sub>/M<sub>n</sub></i>. In
addition, we study the effect of entanglements in ultra-high molecular weight
PE (UHMWPE) on the chain cleavage efficiency during milling by comparing a
disentangled UHMWPE<sup>14</sup>
and an annealed material that contains entanglements. </p>

<h2>EXPERIMENTAL SECTION</h2>

<p class=MsoNormal><b>PE synthesis</b> was carried out in a 500 mL glass
reactor using a magnetic stirrer. The reactor was kept at RT by immersing it in
a water bath. The reactor was purged with nitrogen three times with subsequent
evacuation, while heating it with a heat gun. Subsequently, distilled toluene
(300&nbsp;mL) was introduced and stirred vigorously with a Teflon stir bar. Continuous
and intensive stirring during the reaction is crucial to obtain a PE with a
narrow molecular weight distribution. The reactor was quickly evacuated and
then the ethylene supply was opened. The ethylene pressure was constantly kept
at 1 atm during the whole polymerization. After the solution was saturated with
ethylene for 15&nbsp;min, the polymerization was initiated by fast addition of
a toluene solution (ca. 10 mL) of modified methylaluminoxane (MAO) and the Ti
complex using a syringe, previously prepared in the glovebox. After the desired
reaction time, the ethylene gas feed was stopped and 50&nbsp;mL HCl
(37&nbsp;mol%) were added to terminate the polymerization reaction. The
reaction mixture was left stirring for 60&nbsp;min and then added to acidified
methanol (25&nbsp;ml of concentrated HCl in 500&nbsp;mL of methanol). Solid PE
was recovered by filtration, thoroughly washed with methanol, and dried at 50&nbsp;°C
for 24&nbsp;h in a vacuum oven. Exact synthesis conditions for each PE sample
can be found in <b>Table S1</b>.</p>

<p class=MsoNormal><b>Disentangled PE</b> (PE<sub>single</sub>)
was synthesized according to Schnitte et al.<sup>14</sup>
with a reaction time of 5&nbsp;h using 7.5&nbsp;µmol of C<sub>6</sub>F<sub>13</sub>-tetrasubstituted
N-terphenyl salicylaldiminato Ni polyethylene glycol (PEG, 5 kg/mol) catalyst and 12&nbsp;g of sodium dodecyl sulfate (SDS),
3&nbsp;g of CsOH, 1.5&nbsp;mL toluene as additive, a temperature of 10
&nbsp;&#730;C and a pressure of 40&nbsp;bar yielding
23.95&nbsp;g of polymer.</p>

<p class=MsoNormal><b>Annealing</b> was performed in a septum sealed vial
flushed with nitrogen through a needle, containing 300&nbsp;mg of PE<sub>single</sub>. The vial was placed in a
heating block kept at 135&nbsp;&#730;C for 3&nbsp;h yielding PE<sub>anneal</sub>.</p>

<p class=MsoNormal><b>Ball milling experiments</b> were performed in a Retsch CryoMill
with a 25&nbsp;ml steel jar (Retsch), using five grinding spheres of 10&nbsp;mm
diameter, either made of steel (Retsch) or zirconia (Zhonglong Materials). A
milling frequency of 30&nbsp;Hz was used for all experiments. Experiments
performed for longer than 1&nbsp;h were performed in 1&nbsp;h intervals with
30&nbsp;s pauses. For milling under N<sub>2</sub>, the milling jar was filled
in the glovebox and closed tightly, while for milling under air, the jar was
just filled in the open. Milling at RT denotes milling without cooling. Due to
friction during milling, the milling jar temperature increases up to
45&nbsp;&#730;C. Milling at cryogenic temperatures was performed with the
automatic liquid N<sub>2</sub> dosing system of the device keeping the jar
constantly at -196 °C. The samples were first auto-precooled to the desired low
temperature while shaking with only 5 Hz and then milled T 30 Hz as already
described. For radical trapping experiments, either 30&nbsp;mg di-<i>tert</i>-butyl
hydroxytoluene (BHT) or 21&nbsp;mg 2,2,6,6-tetramethylpiperidinyloxyl (TEMPO)
were added to the grinding jar together with the polymer. </p>

<p class=MsoNormal><b>Differential Scanning Calorimetry (DSC) </b>was performed
on a Netzsch Phoenix 204 F1 at a heating/cooling rate of either 1 or 10&nbsp;&#730;C/min with 4 - 6 mg of polymer in an aluminum pan sealed in the open air. Experiments with 1 °C/min were performed in one heating phase from RT to 180 °C, whereas measurements with 10 °C/min comprised heating to 180 °C, cooling to -50 °C followed by a 2nd heating phase to 180 °C.
Polymer crystallinities were calculated from the heating curves based on a melt enthalpy of 293 J/g for
100% crystalline PE.</p>

<p class=MsoNormal><b>High Temperature Size Exclusion Chromatography (HT-SEC) </b>was
performed on a PolymerChar GPC-IR instrument equipped with PSS Polefin Linear
XL columns (3&nbsp;x&nbsp;30&nbsp;cm), an additional guard column, a four-capillary
viscometer and an IR5 dual wavelength detector (selective for methylene and
methyl units) with 1,2&#8209;dichlorobenzene as solvent with a flow rate of 1&nbsp;ml/min
at 160&nbsp;°C. For samples with <i>M<sub>n</sub></i> &lt; 200 kg/mol a linear calibration with monodisperse PE
standards was employed. UHMWPE and its respective milled samples were measured
with only 0.5 ml/min and evaluated by linear calibration with monodisperse
polystyrene standards.</span></p>

<p class=MsoNormal><b>High Temperature Nuclear Magnetic Resonance Spectroscopy
(HT-NMR) </b>was performed on a Bruker Avance III HD 400 spectrometer (<sup>1</sup>H:
400 MHz) in 1,1,2,2-tetrachloroethane-<i>d<sub>2</sub></i> at 100&nbsp;°C. The
NMR solvent was deoxygenated by bubbling with N<sub>2</sub> for 30 min. All NMR
samples were prepared and sealed in a glovebox to ensure that samples do not
oxidize during the long high temperature NMR measurements. The proton chemical
shifts were referenced to the residual solvent proton signals (C<sub>2</sub>Cl<sub>4</sub>D<sub>2</sub>:
6.00&nbsp;ppm). Chemical group densities are given in mol% with respect to the
total number of CH<sub>2</sub> groups of the polymer backbone.
</p>

<p class=MsoNormal>Upon milling, the polymer chains are cleaved, and the molar
mass distributions continuously shift to lower molar masses (<b>Figure 1a</b>).
In the first 0.5&nbsp;h, a second peak appears at lower mass and over time only
this second peak remains. This causes the <i>M<sub>n</sub></i> (<b>Figure
1b</b>) and weight average molar mass (<i>M<sub>w</sub></i>) to decrease
over time, while the ratio <i>M<sub>w</sub>/M<sub>n</sub></i> increases
(<b>Figure 1c</b>). The limiting molar mass of 2.8 kg/mol, corresponding to a
limiting degree of polymerization of 100, is almost reached for the harshest
milling conditions of cryogenic milling under air with steel grinding spheres.
Milling at RT denotes milling without cooling. Due to friction during milling,
the milling jar temperature increases up to 45&nbsp;&#730;C.</p>

<p class=MsoNormal>Knowing the total weight of polymer used (<i>m</i>=300 mg), the total number of polymer chains in the system and the number of cleavages is calculated (Eq. 1, <b>Figure 1d</b>), where <i>N<sub>A</sub></i> is the Avogadro constant and <i>N<sub>0</sub></i> is the number of chains present in the original polymer sample. The decrease in <i>M<sub>n</sub></i> is faster at shorter milling times and subsequently slows down.</p>

$$
N_t=m/M_n ⋅N_A-N_0	Eq. 1$$



<p class=MsoNormal>Several polymers with different <i>M<sub>n</sub></i> values
were milled at the same conditions (RT, N<sub>2</sub>, ZrO<sub>2</sub>
spheres). All have the same degradation rate, meaning that the initial <i>M<sub>n</sub></i> does not
influence the degradation rate. This is in contrast to what was observed during
ultrasonication,<sup>25</sup>
but in line with observations during ball milling of PS and PMMA, in which the
glass transition temperature of the polymer determined the degradation rate,
rather than <i>M<sub>n</sub></i>.

<p class=MsoNormal><b>Effect of temperature on degradation</b></p>

<p class=MsoNormal>More new chains are formed under cryogenic conditions than
during milling at RT. For example, with PE<sub>160</sub> under N<sub>2</sub> with ZrO<sub>2</sub>
grinding spheres 1.8 times more chains are formed, while under air with steel
grinding spheres 3.9 times more chains are formed (black arrows, <b>Figure 1d</b>).
At cryogenic temperature, the amorphous regions of the PE are glassy. This
could mean that the impact energy dissipates less easily through the polymer,
therefore leading to higher stress concentration causing more cracks and thus
chain cleavage. However, this implies that the chain cleavage probability along
the chain should be different at RT and under cryogenic conditions. This is not
the case as is discussed below. </p>

<p class=MsoNormal>Another effect could be that the lower mobility of the
chains separates radicals spatially and, thus, prevents recombination reactions
(<b>Scheme 2</b>). Radical trapping with BHT leads to the

using steel spheres (<b>Figure&nbsp;S1</b>). The more pronounced effect when adding TEMPO compared to BHT could be
due to the lower trapping efficiency of BHT or the initiation of more chain
cleavage by hydrogen abstraction from PE by TEMPO. Also, kinetically,
recombination is suppressed at lower temperature. That is, its reaction rate
constant is about four orders of magnitude lower at cryogenic conditions (<b>Table
S2</b>).<sup>11</sup>
This allows the chains even more time to separate from each other.

<h3>Cleavage probability is distributed around the center of the chain</h3>

<p class=MsoNormal>To determine the location along the polymer chain, where
cleavage probability is highest, we adapted a model from Glynn et al..<sup>17</sup>
The model predicts the molar mass distribution after a certain number of
cleavage steps based on the assumption that longer chains are cleaved with a
higher probability and that cleavage probability is distributed around the
center of the chain. An extended explanation of the procedure can be found in the SI.</p>

<p class=MsoNormal>In the first step of the model algorithm, certain chains are
selected from the distribution, represented as a probability distribution curve
(<b>Figure&nbsp;S7a</b>). This curve is calculated assuming that the probability
of cleavage increases with molar mass (Eq.&nbsp;2), which is the degree of
polymerization <i>x</i> multiplied
by the monomer molar mass <i>M<sub>0</sub>=28 g/mol</i>. The
dependence is stronger for larger values of the exponent <i>s</i>.
Eq.&nbsp;2 is multiplied with the original molar mass distribution <i>f<sub>n</sub></i> to obtain
the distribution curve of chains to select for cleavage (Eq.&nbsp;3). At the
first simulation step, the molar mass distribution of the original polymer,
obtained from HT-SEC, is used as input.</p>

$$
C=(M_0\cdot x)^s       Eq. 2
$$
$$
P(t,x)=f_{n}⋅C				Eq. 3
$$
<p class=MsoNormal>The script iterates over the polymerization degree <i>x</i> and calculates <i>P(t,x)</i>.
It also calculates how many chains of polymerization degree x are produced by cleavage of chains larger than x, starting from <i>y=x+1</i> (Eq. 4).</p>

$$
∑_{y=x+1}^{\infty}P(t,y)Q(y,x) 				Eq. 4
$$
$$
Q(y,x)=1/(ry√2π)\cdot e^{-((x-y)/2)^2/(2r^2 y^2 )}				Eq. 5
$$

<i>P(t,y)</i> is calculated the same way as <i>P(t,x)</i> (Eq. 3) and <i>Q(y,x)</i> is a Gaussian distribution centered around the polymerization degree <i>y/2</i> (Eq. 5). Here, <i>ry</i> represents the width of the Gaussian distribution, equivalent to what is usually denoted as the standard deviation σ. A larger <i>r</i> represents a cleavage that is more broadly distributed around the center. However, this width of the distribution also increases with the degree of polymerization <i>y</i> (<b>Figure S7b</b>).
At each simulation step <i>t</i>, the process is repeated by iteration over all <i>x</i> and a new chain length distribution is obtained via Eq. 6. Refer to Glynn et al. for the derivation.<sup>17</sup>

$$
f_n (t+1,x)=[(N+t) f_n (t,x)-P(t,x)+2{∑_{y=x+1}^{∞}P(t,y)Q(y,x) }]/(N+t+1)	Eq. 6
$$
Here <i>N</i> represents the number of chains, where in each simulation step a certain fraction of chains is cleaved, roughly scaling with <i>1/(N+t</i>). In our adaptation of the model, r in Eq. 5 is varied together with s in Eq. 2 to fit the measured molar mass distributions after certain milling times.
<p>The fact that the model can fit the molar mass distributions after different milling conditions satisfactorily (<b>Figure 3</b> and <b>Figure S10-27</b>) is a strong indication that the chains are cleaved most likely in the middle of the chain. An even better fit is obtained for the development of <i>M<sub>n</sub></i> and the number of scissions over time. The width of the Gaussian (<i>r=0.10-0.18</i>, <b>Table S3</b>) is similar to what has been observed in ultrasonication of polymer solutions (<i>r=0.15</i>).<sup>17</sup> Smaller values of r are obtained for very high degrees of degradation after 5 h of milling under cryogenic conditions. However, in this case the fits are not reliable (<b>Figures S12, 13, 15, and 24</b>). And when only fitting the first hour of milling under cryogenic conditions, <i>r</i> values in a similar range as for RT milling are obtained (entries 5 and 7 in <b>Table S3</b>).</p>

Since the distribution of chain cleavage probability is intrinsically wider for longer chains (<b>Figure S7b</b>), it is misleading to say that the chain is cleaved “within the middle 15% of the chain” as stated previously for <i>r</i>=0.15.<sup>26</sup> Alternatively, one could state that there is a 68% probability that the chain is cleaved within the inner 30% of the chain. It is a very wide statistical distribution around the center. In addition, the fact that <i>r</i> does not depend on the milling conditions casts doubt on a physical origin but rather suggests that it is unlikely to cleave the polymer at the chain ends under any condition. This is because the chain ends are mobile and a mechanochemical cleavage requires some fixture of the chain at each side around the cleavage location. This is most likely deep in the chain. In this case, the polymer protrudes the longest from either side of the cleavage point and it is most likely that these ends are either fixed in a crystal or an entanglement (<b>Scheme 2a</b>). We will see below that entanglements are crucial to achieve polymer cleavage.
Interestingly, <i>r</i> is dependent on the degree of degradation that is fitted. This is most clearly illustrated when fitting each time point individually for the milling conditions in <b>Figure 3</b>. The fits become much better and <i>r</i> decreases with more cleavage and lower M<sub>n</sub> values (<b>Figure S9</b>). We attempt to explain this by a critical minimum length c that is necessary to immobilize the polymer chain on both sides of the mechanochemical cleavage point, e.g., in crystallites or entanglements (<b>Scheme 4</b>). It is assumed that shorter segments can slip out of their fixation upon mechanical impact, so that the energy of the latter can be dissipated by chain movement without bond cleavage. Mechanochemical chain cleavage is especially unlikely close to dangling chain ends, because these segments are the most flexible.

<p class=MsoNormal><img id="Grafik 4"
src="Figures/Scheme 4.png"></p>
<b>Scheme 4.</b> Illustration of likely cleavage between fixture points, which leads to a statistical cleavage probability centered around the middle of the chain.

In a very simplified model, we simulate the effect of <i>c</i> on <i>r</i> by considering all possible positions of two fixation blocks on polymer chains of a given length <i>l</i>. The location of the two blocks is allowed to freely slide along the polymer chain, without the two blocks overlapping (<b>Figure S28a</b>). For the sake of simplicity, chain cleavage is assumed to occur exactly in the middle between the two blocks. This results in Δ-shaped distributions of cleavage probability centered around the middle of the chain (<b>Figure S28b</b>). The relative width of these distributions (<i>r</i>) correlates with the inverse chain length l (<b>Figure S28c</b>). Thus, shorter chains have a slightly higher selectivity for cleavage deep in the chain, which is in qualitative accordance with the observed lower r values for lower <i>M<sub>n</sub></i> (<b>Figure S9</b>). For the data of <b>Figure S9</b>, a critical minimum fixation length of <i>c</i> = 12.1 ± 2.6 kg/mol was calculated which corresponds to a PE chain segment with a length of 108 ± 23 nm. We recognize that this is quite a bit higher than the limiting molar mass obtained with an extruder previously, which could be due to the relatively mild conditions used for the data in <b>Figure S9</b>, namely milling at RT.

<h3>Effect of entanglement on chain cleavage</h3>
In addition to determining the cleavage location along the chain, we investigated the influence of entanglements on polymer cleavage. We milled an UHMWPE synthesized in a highly diluted aqueous emulsion polymerization. From this polymerization single chain PE (PE<sub>single</sub>) nanocrystals are obtained which cannot entangle during growth because of the compartmentalized character of the reaction and a fast crystallization compared to polymerization rate.<su>14</sup> The analysis of this sample with DSC and the effectiveness in creating entanglements upon annealing (PE<sub>anneal</sub>) is described in the SI, section S6.<sup>27</sup> Upon milling, the degree of crystallinity increased slightly for PEanneal, suggesting that more degradation occurs in the amorphous regions. For PEsingle crystallinity decreased, suggesting that crystalline regions are also affected. While milling leads to some amorphization in the first few minutes, in many of the experiments with different PEs, the degree of crystallinity levels out at around 60 % (<b>Figure S31</b>). Thus, we cannot exclude that some chain cleavage occurs in crystalline regions, however, the amorphous regions seem more affected.
Especially chains spanning over several crystals seem to be prone to degradation. PEanneal experienced a significantly more pronounced decrease in <i>M<sub>n</sub></i> from 1,566 kg/mol to 84 kg/mol than the disentangled, single chain nanocrystal sample, which only decreased to 313 kg/mol (<b>Figure 4</b>). This is a remarkable difference, indicating the effective cleavage of entangled and tie chains.

<h2>CONCLUSIONS</h2>
The synthesis of well-defined PE models with a narrow molar mass distribution allows for the systematic study of chain cleavage mechanisms in polyolefin materials that are much more complex than previously studied fully amorphous PS and PMMA. Polyolefins comprise both amorphous and crystalline domains. Our adaptation of the chain cleavage model by Glynn et al. enables the simultaneous fitting of HT-SEC data of samples obtained after various times of milling. This together with the well-defined materials enables the determination of the chain cleavage location along the chain. Cleavage probability is highest deep inside the chain and lowest at the chain ends. The fitting parameters obtained are very similar for different conditions suggesting a statistical distribution of chain cleavage probability, because the polymer needs to be fixed by entanglements or a crystal on either side of the cleavage site for cleavage to occur and this is the most likely to occur deep inside the polymer chain. The importance of entanglements for cleavage was further illustrated by the significantly faster cleavage of an entangled polymer over its disentangled analog. However, a high crystallinity does not seem to be a prerequisite for cleavage nor is it decisive that the polymer is in a brittle state, i.e., below its glass transition temperature. The similarity in the cleavage probability distribution for milling under cryogenic conditions and at RT suggests that the faster decrease in molar mass observed in the former case is due to slower recombination kinetics at low temperature rather than the lack of stress dissipation in the glassy polymer. In addition, the presence of oxygen during milling leads to a stabilization of radicals and thus a faster decrease in <i>M<sub>n</sub></i>. A similar stabilization effect was observed when milling with steel instead of zirconia spheres. A correlation of degradation rates with molar mass could not be found. These insight provide mechanistic understanding of mechanochemical chain cleavage for PE not previously studied in much depth and expands knowledge previously mostly obtained in ultrasonication or with fully amorphous polymers. The knowledge obtained here on PE is likely also applicable to PP, as both are partially crystalline and have a similar glass transition temperature. As these polymers make up more than 50% of the current global polymer production and are particularly dominant in packaging, the gained insights are useful for designing future mechanochemical strategies for chemical plastic recycling and can also be interesting to better understand unwanted degradation during mechanical recycling via extrusion. Understanding in the ball mill rather than ultrasonication is more industrially relevant as ball milling does not require solvent and is arguably more scalable.

<h3>S5. Chain cleavage modelling</h3>

<p class=MsoNormal>In the first step of the model algorithm, certain chains are
selected from the distribution, represented as a probability distribution curve
(<b>Figure&nbsp;S7a</b>). This curve is calculated assuming that the probability
of cleavage increases with molar mass (Eq.&nbsp;2), which is the degree of
polymerization <i>x</i> multiplied
by the monomer molar mass <i>M<sub>0</sub>=28 g/mol</i>. The
dependence is stronger for larger values of the exponent <i>s</i>.
Eq.&nbsp;2 is multiplied with the original molar mass distribution <i>f<sub>n</sub></i> to obtain
the distribution curve of chains to select for cleavage (Eq.&nbsp;3). At the
first simulation step, the molar mass distribution of the original polymer,
obtained from HT-SEC, is used as input.</p>

$$
C=(M_0\cdot x)^s       Eq. 2
$$
$$
P(t,x)=f_{n}⋅C				Eq. 3
$$
<p class=MsoNormal>The script iterates over the polymerization degree <i>x</i> and calculates <i>P(t,x)</i>.
It also calculates how many chains of polymerization degree x are produced by cleavage of chains larger than x, starting from <i>y=x+1</i> (Eq. 4, <b>Figure S7b</b>).</p>

$$
∑_{y=x+1}^{\infty}P(t,y)Q(y,x) 				Eq. 4
$$
$$
Q(y,x)=1/(ry√2π)\cdot e^{-[((x-y)/2)^2/(2r^2 y^2)]}				Eq. 5
$$

<p><i>P(t,y)</i> is calculated the same way as <i>P(t,x)</i> (Eq. 3) and <i>Q(y,x)</i> is a Gaussian distribution centered around the polymerization degree <i>y/2</i> (Eq. 5). Here, <i>ry</i> represents the width of the Gaussian distribution, equivalent to what is usually denoted as the standard deviation σ. A larger <i>r</i> represents a cleavage that is more broadly distributed around the center. However, this width of the distribution also increases with the degree of polymerization <i>y</i> (<b>Figure S7b</b>). (<b>Figure S3b</b>). <i>Q(y,x)</i> is calculated with <i>x</i> fixed and varying <i>y</i> and then multiplied by <i>P(t,y)</i> to ensure that longer chains are preferably cleavage and taking the chain length distribution at each time step <i>(f<sub>n</sub>(t,x)</i>) into account (<b>Figure S7c</b>).</p>
<p>At each simulation step <i>t</i>, the process is repeated by iteration over all <i>x</i> and a new chain length distribution is obtained via Eq. 6. Refer to Glynn et al. for the derivation.<sup>17</sup></p>

$$
f_n (t+1,x)=[(N+t) f_n (t,x)-P(t,x)+2{∑_{y=x+1}^{∞}P(t,y)Q(y,x) }]/(N+t+1)	Eq. 6
$$

<p>Here <i>N</i> represents the number of chains, where in each simulation step a certain fraction of chains is cleaved, roughly scaling with <i>1/(N+t)</i>. <i>N</i> can be chosen freely as it does not have an influence on the final molar mass distribution. If <i>N</i> is large, the cleavage is milder with each step, while a small <i>N</i> leads to faster cleavage (<b>Figure S8a-d</b>). The resulting molar mass distribution with different choices of <i>N</i> are exemplified in <b>Figure S8e-f</b>.</p>

<p>The chain cleavage probability is modelled with a Gaussian of a certain width <i>ry</i>. In our adaptation of the model, <i>r</i> is varied together with s to fit the measured molar mass distributions after certain milling times. If only one time point were to be fitted, <i>N</i> could simply be varied in addition to <i>r</i> and <i>s</i> to account for the degree of degradation. Thus, <i>N</i> was fixed and the code chose the simulation steps that most closely represented each time point dynamically during the fitting procedure.</p>

<p>At very high degrees of degradation, the model struggles to represent the molar
mass distribution (<b>Figure S12</b>). The model predicts much less cleavage
events with longer simulation times <i>t</i>,
because the fraction of polymers cleaved scales with <i>1/(N+t)</i> (<b>Figure
S29</b>). For example, the first hour of milling is represented by
15&nbsp;steps, while 5&nbsp;h milling requires 6999 instead of the
75&nbsp;steps expected for a linear scaling (<b>Figure&nbsp;3</b>). The
parameter <i>s</i>,
representing the dependence of cleavage probability on molar mass, varies
widely and no trend with number of scissions could be found. This is another
indication that the degradation rate is independent of the initial <i>M<sub>n</sub></i>.</p>

```python
"Figure S29"
"""Note: as this analysis was originally done in matlab and origin. This is just a rough translation to python, which
leads to a slightly different result but is meant to show the approach transparently"""
length = 1000 #longest chain length
ratios = np.arange(0.0, 0.5, 0.05)
ratios[0]=0.01
lengths = 10/ratios
critical=10              #Critical chain length for fixation on one side
iterations=2000000

colors = plt.cm.viridis(np.linspace(0, 1, len(ratios)+2))
fig,ax = plt.subplots(figsize=(3.25, 3.25))
n = 0
for l in lengths:
    cleavage_point=[]
    for i in range(iterations):
        column=np.random.randint(l-2*critical)
        column_1=np.random.randint(l-2*critical)
        if np.any(abs(column-column_1)>0):
            cleavage_point.append((column+column_1)/2+critical+1)
    cleavage_rel=cleavage_point/l
    bins = np.linspace(0, 1, 20)
    histy, binedges = np.histogram(cleavage_rel, bins)
    normhisty = histy/max(histy)
    ax.plot(binedges[1:], normhisty, color=colors[n], label=round(critical/l,2))
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1.1)
    plt.ylabel(r'$\bf{cleavage\ probability}\ \it{(-)}$')
    plt.xlabel(r'$\bf{normalized\ chain\ length}\ \it{(-)}$')
    plt.legend()
    n+=1
plt.show()
```

<b><span style='font-size:14.0pt'>S6. Disentangled and annealed polymer and
crystallinity</span></b></p>

<p class=MsoNormal style='line-height:normal'>We milled an UHMWPE synthesized
in a highly diluted aqueous emulsion polymerization. From this polymerization single
chain PE (PE<sub>single</sub>) nanocrystals are obtained which cannot entangle
during growth because of the compartmentalized character of the reaction and a
fast crystallization compared to polymerization rate.<sup>14</sup> This
polymer, denoted as PE<sub>single</sub>, has a degree of crystallinity of
71.7&nbsp;% and an <i>M<sub>n</sub></i> of
1,566&nbsp;kg/mol. The polymer has a very high melting point of
145.3&nbsp;&#730;C upon first heating at 10&nbsp;&#730;C/min due to the absence
of entanglements (<b>Figure S30a</b>).<sup>27</sup> After the first melting,
entanglements are created which leads to a decreased melting temperature of
136.8&nbsp;&#730;C upon second heating (<b>Figure S30b</b>). We subjected PE<sub>single</sub>
to an annealing step at 135&nbsp;&#730;C for 3&nbsp;h to obtain PE<sub>anneal</sub>,
after which the melting point upon first heating decreased to
133.5&nbsp;&#730;C, demonstrating the presence of entanglements. In addition,
the degree of crystallinity decreased to 53.4&nbsp;%. (<b>Figure S30a</b>).</p>

<p class=MsoNormal style='line-height:normal'>As is expected for a disentangled
polymer, with a slower heating rate of 1&nbsp;&#730;C/min, the melting
temperature of PE<sub>single</sub> decreases to 136.8&nbsp;&#730;C upon first
heating (<b>Figure S30c</b>).<sup>27</sup> With the slower rate, the chains
have more time to unfold. In contrast, an entangled polymer would not show such
a decrease in melting temperature, because the entanglements hinder unfolding.
However, unexpectedly, the annealed sample, denoted as PE<sub>anneal</sub>
melts at an even lower temperature than PE<sub>single</sub> with
1&nbsp;&#730;C/min heating rate. This could be due to a decrease in crystal
size upon annealing for which the decreased crystallinity could be indicative.
Interestingly, the melting temperature after milling increases for both PE<sub>anneal</sub>
and PE<sub>single</sub>.</p>
